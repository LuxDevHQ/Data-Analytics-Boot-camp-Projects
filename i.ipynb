```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weight-height-updated.csv')
```

```python
years = df['Born_Year'].mode()[0]
```

```python
df['Born_Year'].fillna(years, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(30)
```

```python
df.tail(40)
```

```python
def curr_calc():
    born_year = ['2025']
    year1 = []
    for y in born_year:
        years = born_year*10000
        year1.append(years)
    return years
```

```python
current_year = curr_calc()
```

```python
current_year
```

```python
df['Current Year'] = current_year
```

```python
df.head()
```

```python
df['Age'] = df['Current Year'].astype(int) - df['Born_Year'].astype(int)
```

```python
df.head()
```

```python
# w/(h**2)
df['Height']= df['Height']*0.0254
df['Weight']= df['Weight']*0.453592
```

```python
df.head()
```

```python
df['BMI'] = df['Weight']/(df['Height']**2)
```

```python
df.head()
```

```python
def bmi_cat(bmi):
    if bmi < 18.5:
        return "underweight"
    elif 18.5 <= bmi < 24.9:
        return "normal weight"
    elif 25 <= bmi < 29.9:
        return 'over weight'
    else:
        return 'obesse'
```

```python
df['BMI Category'] = df['BMI'].apply(bmi_cat)
```
